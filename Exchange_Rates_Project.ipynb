## Exchange Rates Project

In this project, we are going to do explanatory data visualization on Euro daily exchange rates between 1999 and 2021 dataset.
Our focus in this project will be on the exchange rate between the Euro and the American Dollar.

Dataset can be found [here](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020)

```python
import pandas as pd
```

#### Read the dataset CSV file using pandas

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

#### Do data cleaning

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates.head()
```

**Changes made:**
* We rename the `[US dollar ]` and `Period\Unit:` columns to something easier to type â€” `US_dollar` and `Time`.
* We change the `Time` column to a datetime data type.
* We sort the values by `Time` in ascending order.
* We reset the index (and drop the initial index).

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
```

```python
euro_to_dollar.head()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.head()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar.info()
```

**Changes made:**
* Drop all the rows where the `-` character appears in the `US_dollar` column.
* Convert the `US_dollar` column to a `float` data type.

#### Generate line plot

We will generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

**Observations:**
* If we look at the line's shape, we see many small wiggles â€” rather than seeing a smooth line which shows the visual representation of the daily variation in the exchange rate.
* So, we are going to use the rolling mean instead (also known as the moving average).

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

#### Visualize the data

There are many ways to present the dataset. This time, we will show how the euro-dollar rate has changed during the coronavirus pandemic. We will show the 2020 data and the 2016-2019 data as a baseline using line plot.

```python
corona = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year == 2020)]
baseline = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2016) & (euro_to_dollar['Time'].dt.year <= 2019)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(baseline['Time'], baseline['rolling_mean'], linewidth=1, color='#348feb')
ax.plot(corona['Time'], corona['rolling_mean'], linewidth=3, color='#e23d28')

ax.set_xticklabels([])
ax.set_yticklabels([])

x = 16760.0
for year in ['2016', '2017', '2018', '2019', '2020', '2021']:
    ax.text(x, 1.035, year, alpha=0.5, fontsize=11)
    x += 365

y = 1.048
for rate in ['1.05', '1.1', '1.15', '1.2']:
    ax.text(16640.0, y, rate, alpha=0.5, fontsize=11)
    y += 0.05

ax.text(16600.0, 1.28, 'EURO-USD rate quickly recovered during corona pandemic crisis',
         fontsize=20, weight='bold')
ax.text(16600.0, 1.266, 'EURO-USD exchange rate almost back to its highest value in 2018 within one year', fontsize=16)

ax.text(16600.0, 1.01, 'Â©DATAQUEST' + ' '*110 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)

ax.axhspan(ymin=1.2, ymax=1.25, xmin= 0.38, alpha=0.3, color='grey')

ax.grid(alpha=0.5)
plt.show()
```
